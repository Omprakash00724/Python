```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('h1b.csv')
df
```

```python
df.info()
```

<h1>DATA CLEANSING<h1>

```python
df.fillna(0, inplace=True)
```

```python
df.replace({"Y": 1, "N": 0}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
#Petition distributed by case statuses
petition_counts = df['CASE_STATUS'].value_counts()
print(petition_counts)
```

```python
plt.figure(figsize=(10, 9))
plt.pie(petition_counts, labels=petition_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Petitions by Case Statuses")
plt.legend(title="Case Statuses", loc="best")
plt.show()
```

```python
#Top employers filing the petitions
top_employers = df['EMPLOYER_NAME'].value_counts().head(5)
top_employers
```

```python
#Top 15 companies that filed the most visa application
top_companies = df['EMPLOYER_NAME'].value_counts().head(15)
top_companies
```

```python
#Top employers granting maximum prevailing wages
top_employers = df.groupby('EMPLOYER_NAME')['PREVAILING_WAGE'].max()
top_employers_sorted = top_employers.sort_values(ascending=False)
top_10_employers = top_employers_sorted.head(10)
print(top_10_employers)
```

```python
##Top 20 states filing highest visa petitions
df['STATE'] = df['WORKSITE'].str.split(',').str[1]
top_states = df['STATE'].value_counts().head(20)
print(top_states)
```

```python
#Petition distribution by year
petition_distribution = df['YEAR'].value_counts().sort_index()
print(petition_distribution)
```

```python
#Number of visa petitions denied
denied_petitions = df[df['CASE_STATUS'] == 'DENIED']
num_denied_petitions = len(denied_petitions)
print("Number of visa petitions denied:", num_denied_petitions)
```

```python
#Rate of visas denied from last 5 years
denied_petitions_by_year = denied_petitions['YEAR'].value_counts()

denial_rate_by_year = pd.DataFrame({
    'Total Petitions': petition_distribution,
    'Denied Petitions': denied_petitions_by_year,
})

denial_rate_by_year['Denial Rate (%)'] = (denial_rate_by_year['Denied Petitions'] / denial_rate_by_year['Total Petitions']) * 100

last_5_years = denial_rate_by_year.sort_index().tail(5)

print(last_5_years)
```

```python
# Number of visa petitions filed by states
state_petitions=df['WORKSITE'].str.split(',').str[1].value_counts()
state_petitions
```

```python
# Rate of denial of visa petition by the states
newdf = df.copy()

newdf['STATE'] = newdf['WORKSITE'].str.split(',', expand=True)[1].str.strip()
grouped_by_state = newdf.groupby('STATE')['CASE_STATUS'].value_counts().unstack().fillna(0)
grouped_by_state['Denial Rate (%)'] = (grouped_by_state['DENIED'] / grouped_by_state.sum(axis=1)) * 100
grouped_by_state_sorted = grouped_by_state.sort_values('Denial Rate (%)', ascending=False)
print(grouped_by_state_sorted)
```

```python
#Top 25 job titles
df['JOB_TITLE'].value_counts()[0:25]
```

```python
#Top 10 occupations with the highest wages
df.sort_values('PREVAILING_WAGE',ascending=False)[['JOB_TITLE','PREVAILING_WAGE']][0:10].reset_index()
```
